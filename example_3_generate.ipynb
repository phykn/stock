```python
import numpy as np
import matplotlib.pyplot as plt
from src.stock import to_ror
from src.generate import generate
from src.fit import cauchy_fit
```

```python
x0s = np.linspace(-1, 1, num = 100)
gammas = np.linspace(0.1, 2, num = 100)

p_x0s = []
p_gammas = []

for x0, gamma in zip(x0s, gammas):
    x = generate(x0 = x0, gamma = gamma)
    ror = to_ror(x)
    p = cauchy_fit(ror)

    p_x0s.append(p['x0'])
    p_gammas.append(p['gamma'])

plt.figure(figsize = (10, 4))

plt.subplot(1, 2, 1)
plt.scatter(x0s, p_x0s, color = 'k', s = 15)
plt.xlabel('x0 (GT)')
plt.ylabel('x0 (PRED)')
plt.xlim(-1, 1)
plt.ylim(-1, 1)

plt.subplot(1, 2, 2)
plt.scatter(gammas, p_gammas, color = 'k', s = 15)
plt.xlabel('gamma (GT)')
plt.ylabel('gamma (PRED)')
plt.xlim(0.1, 2)
plt.ylim(0.1, 2)

plt.tight_layout()
plt.show()
```
